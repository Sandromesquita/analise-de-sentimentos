```python
stopwords_list = nltk.corpus.stopwords.words('portuguese')
np.transpose(stopwords_list)
```

```python
stopwords_list.append('tipo')
stopwords_list.append('tão')
stopwords_list.append('tudo')
stopwords_list.append('vai')
```

```python
print(stopwords_list)
```

```python
def removeStopWords(text):
    phrases = []
    for (words, sentiment) in text:
        #Cria-se uma lista para extrair apenas as palavras que não estão na stopwords_list
        without_stop = [p for p in words.split() if p not in stopwords_list]
        #Inserindo as frases com as labels (sentimento) já tratadas pela stopwords_list
        phrases.append((without, sentiment))
    return phrases
```

```python
def applyStemmer(text):
    stemmer = nltk.stem.RSLPStemmer()
    #Escolhido o RSLPStemmer pois é específico da lingua portuguesa
    phrases_without_stemming = []
    for (words, sentiment) in text:
        with_stemming = [str(stemmer.stem(p)) for p in words.split() if p not in stopwords_list]
        phrases_without_stemming.append((with_stemming, sentiment))
    return phrases_without_stemming
```

Cria-se uma função para retornar apenas as palavras da frase, sem a classificação (sentimento)

```python
def searchWords(phrases):
    all_words = []
    for (words, sentiment) in phrases:
        all_words.extend(words)
    return all_words
```

Função que verifica a quantidade de vezes que a palavras é mencionada

```python
def searchFrequency(words):
    words = nltk.FreqDist(words)
    return words
```

Função que retorna somente as palaras unicas

```python
def searchUniqueWords(frequency):
    freq = frequency.keys()
    return freq

unique_training_words = searchUniqueWords(training_frequency)
unique_test_words = searchUniqueWords(test_frequency)
```

Função para identificar quais palavras únicas estão no documento passado para a função

```python
def wordExtractor(document):
    #Utilizado o set() para associar a variável doc com o parâmetro que está chegando
    doc = set(document)
    characteristics = {}
    for words in unique_training_words:
        characteristics['%s' % words] = (words in doc)
    return characteristics
```

```python
def test_words_extractor(document):
    doc = set(document)
    characteristics = {}
    for words in unique_test_words:
        characteristics['%s' % words] = (words in doc)
    return characteristics
```
